```python
CI1 = []
CI2 = []
```

```python
import numpy as np
import scipy 
stats.index
for i in stats.index:
    m, s, n = stats.loc[i]
    x=scipy.stats.t.interval(.95, n-1, m,s/np.sqrt(n-1))
    CI1.append(x[1])
    CI2.append(x[0])
CI1
```
